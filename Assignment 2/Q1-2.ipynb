```python
import numpy as np
import matplotlib.pyplot as plt

n1 = 1.5
n2 = 1.47
coreThickness = 15 * 10**(-6)  # This is supposed to be 7.5?
wavelength = 1.55 * 10**(-6)

k = (2 * np.pi)/wavelength
thetas = np.linspace(0, np.pi/2, 10000)

te = lambda x: 2 * np.arctan( np.sqrt(np.sin(x)**2 - (n2/n1)**2)/np.cos(x)) - k * n1 * coreThickness * np.cos(x)
tm = lambda x: 2 * np.arctan( np.sqrt(np.sin(x)**2 - (n2/n1)**2)/(np.cos(x) * (n2/n1)**2 )) - k * n1 * coreThickness * np.cos(x)

TEs = te(thetas)
TMs = tm(thetas)

index = 0
for i in TEs:
    if (abs(i) < 10**(-2)):
        print((thetas[index] * 180)/np.pi)
    index +=1

index = 0
for i in TMs:
    if (abs(i) < 10**(-2)):
        print((thetas[index] * 180)/np.pi)
    index +=1

# First Graph for TE
plt.figure()  # Create a new figure
plt.title("Fundamental TE")
plt.plot((thetas * 180)/np.pi, TEs, label="TE")
plt.xlabel("Theta (Radians)")
plt.ylabel("Vals")
plt.legend()
plt.show()

# First Graph for TE
plt.figure()  # Create a new figure
plt.title("Fundamental TE")
plt.plot((thetas * 180)/np.pi, TMs, label="TE")
plt.xlabel("Theta (Radians)")
plt.ylabel("Vals")
plt.legend()
plt.show()
```
