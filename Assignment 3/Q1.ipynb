```python
#Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# Linspace of X
x = np.linspace(0, 10, 1000000)

# Set the C constant to 1
C = 1

# Lambda function for the Function n_x(x)
nx = lambda x: C*np.sqrt(x)*np.exp(-x)

# Get the y Values
densityCurve = nx(x)

# Get Maximum Value and the Half Max
maxVal = max(densityCurve)
halfMax = maxVal / 2

# Max Solving EQ
nx_max = lambda x: C*np.sqrt(x)*np.exp(-x) - maxVal

# Half Max Solving EQ
nx_halfmax = lambda x: C*np.sqrt(x)*np.exp(-x) - halfMax

# Solve for root of Max
rootMax = fsolve(nx_max, 2)

# Get Half Maxes
rootHalfMax1 = fsolve(nx_halfmax, 0.1)
rootHalfMax2 = fsolve(nx_halfmax, 2)

# Get FWHM
FWHM = rootHalfMax2[0] - rootHalfMax1[0]

# Display Data
print(f"Maximum = {maxVal} @ X = {rootMax[0]}")
print(f"Full Width Half Maximum = {FWHM}")

# Plot data
plt.plot(x, densityCurve, label="C=1")
plt.title("Plot Density of Electrons in the Conduction Band in terms of X")
plt.xlabel("Value of X")
plt.ylabel("Relative Density of Electrons")
plt.savefig("Density.png")
plt.show()

# Part c (Numerical Integration)
x = np.linspace(0, 10, 1000000)
dx = x[1] - x[0]

# Integral of Nx
NxInt = 0
for xVal in x:
    Area = dx * nx(xVal)
    NxInt += Area
    
# Integral of x*Nx
XNxInt = 0
for xVal in x:
    Area = dx * xVal * nx(xVal)
    XNxInt += Area
    
Average = XNxInt / NxInt

print(f"X_Average = {Average}")
```
